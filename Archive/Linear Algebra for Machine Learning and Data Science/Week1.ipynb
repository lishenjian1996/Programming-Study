# Linear dependence and independence

# Singular vs non-singular matrices


(1+4-2)-(-1+4+2)=3-5=-2

```python
import numpy as np
A = np.array([
        [1, 2, 3],
        [0, 2, 2],
        [1, 4, 5]
    ], dtype=np.dtype(float))
d = np.linalg.det(A)

print(f"Determinant of matrix A: {d:.2f}")
```
