# Portfolio Construction with Time-Varying Risk Parameters


Q: Consider the following stream of returns: +1%, -2%, -1%, +2%. What are the corresponding (arithmetic) average return and volatility estimates? 

Average return is (1%-2%-1%+2%)/4=0%. Variance of returns is (1%2-2%2-1%2+2%2)/4=0.025%. Volatility is square-toot of variance: √0.025%=1.58%.

Q: How many parameters do you need to estimate when using a 2-factor models with GARCH(1,1) model for the volatility of each one of the two factors?

We first need 500 volatility estimates for individual stock returns, plus 500 estimates of betas of stocks with respect to factor 1, 500 estimates of betas of stocks with respect to factor 2, and finally 3 GARCH parameter estimates for each factor, which gives a total of 500+500+500+2x3=1,506, which is not much more than if we had assumed constant volatility parameters.
