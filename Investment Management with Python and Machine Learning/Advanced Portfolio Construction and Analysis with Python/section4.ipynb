# Portfolio Construction with Time-Varying Risk Parameters


Q: Consider the following stream of returns: +1%, -2%, -1%, +2%. What are the corresponding (arithmetic) average return and volatility estimates? 

Average return is (1%-2%-1%+2%)/4=0%. Variance of returns is (1%2-2%2-1%2+2%2)/4=0.025%. Volatility is square-toot of variance: √0.025%=1.58%.
