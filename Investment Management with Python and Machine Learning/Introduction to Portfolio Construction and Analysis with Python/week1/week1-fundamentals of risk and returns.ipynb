```python
def greet(say, person):
    """
    Generates a string that is a greeting to the specified person
    Parameters:
        say (str): What do you want to say?
        person (str): Who do you want to say it to?
    Returns:
        Greetings, which is a string
    """
    return f'{say} {person}!'
```

# Fundamentals of Returns
Price Return: (Pt+1 - Pt)/Pt
Total Return: (Pt+1 + Dt,t+1 - Pt)/Pt

Multi Period Returns: Rt,t+2 = (1+Rt,t+1)(1+Rt+1,t+2)-1
e.g. buy a returns 10% first day, lose -3% second day, then (1+0.1)(1-0.03)-1=0.067

# Measures of Risk and Reward

We can't compare the volatility from Daily data with the volatility from Monthly data
daily data 0.001 x square root of 252 = yearly data 1.52%

Lab Session-Risk Adjusted returns

```python
import pandas as pd
prices=pd.read_csv('sample_prices.csv')
returns=prices.pct_change()
returns
```

```python
#delete N/A
returns=returns.dropna()
returns
```

```python
returns.std()
```

```python
deviations=returns-returns.mean()
squared_deviations=deviations**2
variance=squared_deviations.mean()
import numpy as np
volatility=np.sqrt(variance)
volatility
```

```python
number_of_obs=returns.shape[0]
variance=squared_deviations.sum()/(number_of_obs-1)
volatility = variance**0.5
```

```python
returns=pd.read_csv('Portfolios_Formed_on_ME_monthly_EW.csv',header=0,index_col=0, parse_dates=True, na_values=-99.99)
```

```python
returns.head()
```

```python
columns=['Lo 10','Hi 10']
returns=returns[columns]
returns.head()
```

```python
returns=returns/100
returns.head()
```

```python
returns.columns=['SmallCap', 'LargeCap']
returns.head()
```

```python
returns.plot.line()
```

```python
returns.std()
```

```python
annualized_vol=returns.std()*np.sqrt(12)
annualized_vol
```

```python
returns
```

```python
n_months=returns.shape[0]
#return over one month
return_per_month=(returns+1).prod()**(1/n_months)-1
return_per_month
```

```python
annualized_vol
```

```python
annualized_return=(return_per_month+1)**12-1
annualized_return
```

```python
annualized_return=(returns+1).prod()**(12/n_months)-1
annualized_return
```

```python

annualized_return/annualized_vol
```

```python
# assume risk free rate is 3%
# Sharpe ratio
riskfree_rate=0.03
excess_return=annualized_return-riskfree_rate
sharpe_ratio=excess_return/annualized_vol
sharpe_ratio
```

# Max Drawdown

The maxmium loss from the previous high to a subsequent low.

Buy as its highest value and sold at the bottom

The worst case.

Look at the Prior Peak at any point in time.

Calmar ratio is defined as the Ratio of the annualized return over the trailing 36 months to the Maximum Drawdown over those trailing 36 months.
