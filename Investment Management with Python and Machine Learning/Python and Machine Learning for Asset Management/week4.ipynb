# Introduction to economic regimes


Regime is a period of time in which the characteristics of market performance and related economic activity are reasonably stable.

Question: 

What identifies an economic regime?

Answer:

Relatively homogeneous patterns in economic conditions (returns and risks), Relatively different behavior across regimes.


We have shown from historical evidence that crash periods experience much different performance behavior than other economic time periods.  If this conjecture is true, we must categorize time periods by differing market performance, such as severe losses in risky assets during crash periods.  Machine learning algorithms can assist in the classification task (including the trend filtering algorithm).


Question:

What are the primary motivations for a multi-regime framework?

Answer:
1. A more accurate representation of the tail risks
2. Incorporates dynamic changes in an economy
3. Improves performance

There is much empirical evidence that crash periods display performance different from normal (non-crash) periods.  During crashes, volatility and correlations increase, often to high elevated  levels, as compared with normal conditions.  The degree of diversification becomes much more limited.  At the same time, investors transfer capital to so-called safe assets such as government bonds. This tendency is called: a flight-to-quality, and as a consequence the correlation of returns between risky assets such as stocks and government bonds switches from modestly positive to negative.  A multi-period framework has greater capability to depict these dynamic patterns than is possible with a single-period system and is more realistic. 

 Itâ€™s natural that a realistic risk management system contributes to improved future performance.  For instance, an investor will be prepared for a variety of future events and be able to develop plans (policy rules) that protect her during crash periods.  As the saying goes: forewarned is forearmed.

# Trend filtering


A multi-regime model can be applied to a risk management system with several time steps by adding a process for determining the transitions between regimes across time periods. Thus, if the current economic state at time t is a normal regime, we can calculate the next periodâ€™s state (t+1) by means of a Markov transition matrix.  Thereby in general, each scenario will consist of both normal and crash regimes, as realizations of the Markov generation process.  As we recall, the scenario set depict representative and random samples for the entire planning period.

Trend filtering is a strictly Non-parametric method for indentifying economic regimes over historical time periods.

# A two regime portfolio example

Question: 

What are the ingredients for constructing a single-period portfolio model with a blend of two regimes?


Answer: 

1. Sample scenarios from two (multi-normal) distributions. The estimates for each regime are carried out by reference to the three aforementioned considerations. 

2. Estimating expected returns of hybrid securities via factor based approaches.

3. A scenario based optimization model, along with appropriate reward and risk measures.

4. Probability estimates of the transition between regimes between time t and t+1.

We have shown that a scenario version of a single period portfolio model provides a workable format for modeling both single and multiple period regimes.  For instance, suppose that we estimate that there is a 10 percent chance of a contraction over the next year. We could construct, for example, a scenario system with 100 scenarios depicting contraction conditions, and 900 scenarios taken from the normal conditions.
