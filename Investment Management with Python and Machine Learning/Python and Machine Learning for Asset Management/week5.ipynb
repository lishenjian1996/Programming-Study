Question: 

Why do we employ a logistic regression rather than the traditional ordinary least square regression when estimating the probability of a crash?   

Answer：

We label historical periods as normal or crash and logistic regression is designed to address binary outcomes.


Explaination:

Traditional OLS regression is designed to predict dependent variables with continuous outcomes. In contrast, logistic regression is suited to outcomes possessing binary and other discrete properties.  In our studies, we identify crash periods  via several approaches:  pure judgment, hidden Markov models,  trend filtering, structural breaks.  Our supposition, confirmed with much historical evidence, is that crash periods display unique patterns, as compared with normal economic periods.

Question: 

Can a regularized logistic regression reduce the number of non-zero beta coefficients? 

Answer: 

Yes


Eplaination: 

The penalty term in the regularization provides a mechanism for shrinking the beta coefficients, and similar to the LASSO, the number of non-zero beta coefficients will be subsequently reduced. This reduction can be helpful in identifying the most relevant features/factors.   Note that the logistic regression requires a convex solver, which limits the feasible size of the analysis.
