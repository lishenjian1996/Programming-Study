Gradient Descent

In order to minimise J(w,b)

take baby step to move down hill
every step will choose as different starting points
just in order to find the nearst way to move down the hill




#Inplementing gradiend descent algorithm

## w=w-alpha(d/dw)J(w,b)

=is the assignment operator
== is the equal sign

Alpha is 'Learning rate', the larger learning rate, the more aggressive gradiend descent procedure (which mean move down the hill with larger steps), vice versa.

d/dw J(w,b) is derivative term of the cost function J. It indicates which direction you want to take your baby step.

## b=b-alpha(d/db)J(w,b)

Repeat these two update steps until the 'algorithm converges'
which mean it reached the point at a local minimum where the 'w' and 'b' no longer change much with each additional step.

Note: the w=w-alpha(d/dw)J(w,b), and b=b-alpha(d/db)J(w,b) need to be updated at the same time (simulataneously).




Gradient descent intuition

slope>0
w=w-alpha*(positive number) it means adding a negative number to the left in order to find the minimum 'w'

slope<0
w=w-alpha*(negative number) it means adding a positive number to the right in order to find the minimum 'w'
